```python
"""
=============================================================================
  🩺  Medical Image Analysis: Breast Cancer Detection (Colab Edition) v9.0
=============================================================================

  Complete Bug-Fix History:
      v3 ✅ Removed rescale=1./255 (broke EfficientNet → AUC = 0.5)
      v3 ✅ Preprocessing moved inside model as a dedicated layer
      v3 ✅ Grad-CAM rebuilt with Functional API (fixed first AttributeError)
      v3 ✅ Only unfreeze last 30 layers in phase 2 (fixed loss spike 1.2→1.7)
      v3 ✅ BatchNorm layers kept frozen during fine-tuning
      v4 ✅ Separate val_datagen with no augmentation
      v4 ✅ EarlyStopping + ModelCheckpoint + ReduceLROnPlateau + TensorBoard
      v4 ✅ Auto-detects last conv layer name (no hardcoding)
      v5 ✅ Lambda(preprocess_input) → registered custom Layer (fixed save/load)
      v6 ✅ keras.saving.* → tf.keras.utils.* (TF 2.x compatibility)
      v7 ✅ Grad-CAM ValueError "No such layer: top_conv"
      v8 ✅ Grad-CAM gradient was None (two disconnected forward passes)
      v9 ✅ Grad-CAM KeyError on tensor ID (cross-graph boundary violation)
             Root cause: after load_model(), the nested efficientnetb0
             sub-model lives in its OWN graph scope. Accessing
             sub_model.get_layer('top_conv').output gives a tensor node
             from the sub-model's internal graph. Using that tensor as
             an output of a NEW Model(...) built from full_model.inputs
             crosses graph boundaries — Keras cannot trace the path
             from full_model's input through the sub-model's internal
             tensor, causing KeyError on the tensor ID lookup.
             Fix: use a forward-hook pattern. Register a Keras callback
             via a custom layer that captures the conv output tensor
             DURING the forward pass of full_model itself. No new Model
             is constructed — no graph boundaries are crossed.

  Results (v8 run):
      Phase 1 best val_auc : 0.9640
      Phase 2 best val_auc : 0.9643
      Accuracy : 85.4%  |  Malignant AUC : 0.9282

  Setup:
      1. Runtime → Change runtime type → GPU (T4)
      2. Paste Kaggle credentials in STEP 1
      3. Runtime → Run all

=============================================================================
"""

# =============================================================================
# IMPORTS
# =============================================================================

import os, sys, warnings, zipfile
warnings.filterwarnings('ignore')

import cv2
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.applications.efficientnet import preprocess_input
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import (
    EarlyStopping, ModelCheckpoint, ReduceLROnPlateau, TensorBoard
)
from sklearn.metrics import (
    classification_report, confusion_matrix,
    roc_auc_score, roc_curve
)
from sklearn.utils.class_weight import compute_class_weight

# ── Reproducibility ───────────────────────────────────────────────────────────
SEED = 42
tf.random.set_seed(SEED)
np.random.seed(SEED)
os.environ['PYTHONHASHSEED'] = str(SEED)

print("=" * 60)
print("  🩺  Breast Cancer Detection  |  Colab Edition  v9.0")
print("=" * 60)
print(f"  TensorFlow : {tf.__version__}")
print(f"  Python     : {sys.version.split()[0]}")
gpus = tf.config.list_physical_devices('GPU')
print(f"  GPU(s)     : {[g.name for g in gpus] if gpus else 'None — switch to GPU!'}")
print("=" * 60)


# =============================================================================
# CUSTOM PREPROCESSING LAYER
# =============================================================================
# @tf.keras.utils.register_keras_serializable → saves/loads correctly in TF 2.x

@tf.keras.utils.register_keras_serializable(package='Custom', name='EfficientNetPreprocess')
class EfficientNetPreprocess(layers.Layer):
    """Serialisable EfficientNetB0 preprocessing: [0,255] → [-1,1]."""
    def call(self, inputs):
        return preprocess_input(inputs)
    def get_config(self):
        return super().get_config()


# =============================================================================
# CONFIGURATION
# =============================================================================

DATA_DIR         = "dataset/Dataset_BUSI_with_GT"
CLASS_NAMES      = ['benign', 'malignant', 'normal']
NUM_CLASSES      = len(CLASS_NAMES)

IMG_SIZE         = (224, 224)
BATCH_SIZE       = 16

PHASE1_EPOCHS    = 20
PHASE2_EPOCHS    = 15
PHASE1_LR        = 3e-4
PHASE2_LR        = 2e-5
UNFREEZE_LAST    = 30
DROPOUT_RATE     = 0.4
LABEL_SMOOTHING  = 0.1
VAL_SPLIT        = 0.2

BEST_MODEL_PATH  = "best_model.keras"
FINAL_MODEL_PATH = "final_model.keras"


# =============================================================================
# STEP 1: DATASET DOWNLOAD
# =============================================================================

os.environ['KAGGLE_USERNAME'] = "99"   # ← paste here
os.environ['KAGGLE_KEY']      = "KGAT_997216274703023547f6a211a24edf19"    # ← paste here


def download_dataset() -> bool:
    if os.path.isdir(DATA_DIR):
        existing = [
            c for c in CLASS_NAMES
            if os.path.isdir(os.path.join(DATA_DIR, c))
            and len(os.listdir(os.path.join(DATA_DIR, c))) > 0
        ]
        if existing:
            total = sum(len(os.listdir(os.path.join(DATA_DIR, c))) for c in existing)
            print(f"  ✅  Dataset already present ({total} images). Skipping download.")
            return True

    print("\n[STEP 1] Connecting to Kaggle and downloading dataset …")
    os.system("pip install -q kaggle")

    if os.environ.get('KAGGLE_USERNAME', '') == "YOUR_KAGGLE_USERNAME":
        print("  ❌  Paste your Kaggle credentials in STEP 1.")
        return False

    ret = os.system(
        "kaggle datasets download -d aryashah2k/breast-ultrasound-images-dataset"
    )
    if ret != 0:
        print("  ❌  Download failed.")
        return False

    zips = [f for f in os.listdir('.') if f.endswith('.zip')]
    if not zips:
        print("  ❌  ZIP not found after download.")
        return False

    print(f"  📦  Extracting {zips[0]} …")
    with zipfile.ZipFile(zips[0], 'r') as zf:
        zf.extractall("dataset")
    os.remove(zips[0])
    print("  ✅  Dataset downloaded and extracted successfully!")
    return True


if not download_dataset():
    raise SystemExit("Dataset unavailable. Fix credentials and re-run.")


# =============================================================================
# STEP 2: DATA GENERATORS
# =============================================================================

print("\n[STEP 2] Loading and preparing images …")

if not os.path.exists(DATA_DIR):
    raise SystemExit(f"❌  Folder not found: '{DATA_DIR}'")

train_datagen = ImageDataGenerator(
    rotation_range=15, width_shift_range=0.10,
    height_shift_range=0.10, horizontal_flip=True,
    zoom_range=0.15, brightness_range=[0.85, 1.15],
    fill_mode='nearest', validation_split=VAL_SPLIT,
)
val_datagen = ImageDataGenerator(validation_split=VAL_SPLIT)

train_generator = train_datagen.flow_from_directory(
    DATA_DIR, target_size=IMG_SIZE, batch_size=BATCH_SIZE,
    class_mode='categorical', classes=CLASS_NAMES,
    subset='training', shuffle=True, seed=SEED,
)
val_generator = val_datagen.flow_from_directory(
    DATA_DIR, target_size=IMG_SIZE, batch_size=BATCH_SIZE,
    class_mode='categorical', classes=CLASS_NAMES,
    subset='validation', shuffle=False, seed=SEED,
)

print(f"\n  Training   : {train_generator.samples} images | {len(train_generator)} steps/epoch")
print(f"  Validation : {val_generator.samples} images  | {len(val_generator)} steps/epoch")
print(f"  Class map  : {train_generator.class_indices}")


# =============================================================================
# STEP 3: CLASS WEIGHTS
# =============================================================================

print("\n[STEP 3] Computing class weights to fix data imbalance …")

weights_arr = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(train_generator.classes),
    y=train_generator.classes,
)
class_weight_dict = dict(enumerate(weights_arr))

print(f"\n  {'Class':<12} {'Samples':>8}  {'Weight':>8}  {'Note'}")
print(f"  {'─' * 55}")
for i, (cls, w) in enumerate(zip(CLASS_NAMES, weights_arr)):
    n    = int(np.sum(train_generator.classes == i))
    note = "← pay extra attention" if w > 1.0 else ""
    print(f"  {cls:<12} {n:>8}  {w:>8.4f}  {note}")
print("\n  ℹ️   Weight > 1.0 = model penalised more for missing that class.")


# =============================================================================
# STEP 4: BUILD THE MODEL
# =============================================================================

print("\n[STEP 4] Building Transfer Learning Model …")

base_model = EfficientNetB0(
    include_top=False, weights='imagenet', input_shape=(*IMG_SIZE, 3)
)
base_model.trainable = False

inputs  = tf.keras.Input(shape=(*IMG_SIZE, 3), name='image_input')
x       = EfficientNetPreprocess(name='efficientnet_preprocess')(inputs)
x       = base_model(x, training=False)
x       = layers.GlobalAveragePooling2D(name='gap')(x)
x       = layers.Dense(512, name='fc_512')(x)
x       = layers.BatchNormalization(name='bn_1')(x)
x       = layers.Activation('relu', name='relu_1')(x)
x       = layers.Dropout(DROPOUT_RATE, name='dropout_1')(x)
x       = layers.Dense(256, name='fc_256')(x)
x       = layers.BatchNormalization(name='bn_2')(x)
x       = layers.Activation('relu', name='relu_2')(x)
x       = layers.Dropout(DROPOUT_RATE * 0.6, name='dropout_2')(x)
outputs = layers.Dense(NUM_CLASSES, activation='softmax', name='predictions')(x)

model = tf.keras.Model(inputs, outputs, name='BreastCancer_EfficientNetB0_v9')

model.compile(
    optimizer = tf.keras.optimizers.Adam(learning_rate=PHASE1_LR),
    loss      = tf.keras.losses.CategoricalCrossentropy(label_smoothing=LABEL_SMOOTHING),
    metrics   = ['accuracy', tf.keras.metrics.AUC(name='auc'),
                 tf.keras.metrics.Precision(name='precision'),
                 tf.keras.metrics.Recall(name='recall')],
)

model.summary()
print(f"\n  Total params     : {model.count_params():,}")
print(f"  Trainable params : {sum(tf.size(v).numpy() for v in model.trainable_variables):,}  (head only)")


# =============================================================================
# CALLBACKS
# =============================================================================

os.makedirs('logs', exist_ok=True)

def make_callbacks(phase: int) -> list:
    return [
        EarlyStopping(monitor='val_auc', patience=6,
                      restore_best_weights=True, mode='max', verbose=1),
        ModelCheckpoint(filepath=BEST_MODEL_PATH, monitor='val_auc',
                        save_best_only=True, mode='max', verbose=1),
        ReduceLROnPlateau(monitor='val_loss', factor=0.5,
                          patience=4, min_lr=1e-8, verbose=1),
        TensorBoard(log_dir=f'logs/phase{phase}', histogram_freq=0),
    ]


# =============================================================================
# STEP 5: TRAIN
# =============================================================================

print(f"\n--- PHASE 1: Training Medical Head (Base Frozen) — max {PHASE1_EPOCHS} epochs ---")

history_phase1 = model.fit(
    train_generator, validation_data=val_generator,
    epochs=PHASE1_EPOCHS, class_weight=class_weight_dict,
    callbacks=make_callbacks(1), verbose=1,
)
best_p1 = max(history_phase1.history['val_auc'])
print(f"\n  ✅  Phase 1 complete — best val_auc: {best_p1:.4f}")

print(f"\n--- PHASE 2: Fine-Tuning Last {UNFREEZE_LAST} Base Layers — max {PHASE2_EPOCHS} epochs ---")

base_model.trainable = True
for layer in base_model.layers[:-UNFREEZE_LAST]:
    layer.trainable = False
for layer in base_model.layers:
    if isinstance(layer, layers.BatchNormalization):
        layer.trainable = False

n_unfrozen = sum(1 for l in base_model.layers if l.trainable)
print(f"  Base layers unfrozen : {n_unfrozen} / {len(base_model.layers)}")

model.compile(
    optimizer = tf.keras.optimizers.Adam(learning_rate=PHASE2_LR),
    loss      = tf.keras.losses.CategoricalCrossentropy(label_smoothing=LABEL_SMOOTHING),
    metrics   = ['accuracy', tf.keras.metrics.AUC(name='auc'),
                 tf.keras.metrics.Precision(name='precision'),
                 tf.keras.metrics.Recall(name='recall')],
)

history_phase2 = model.fit(
    train_generator, validation_data=val_generator,
    epochs=PHASE2_EPOCHS, class_weight=class_weight_dict,
    callbacks=make_callbacks(2), verbose=1,
)
best_p2 = max(history_phase2.history['val_auc'])
print(f"\n  ✅  Phase 2 complete — best val_auc: {best_p2:.4f}")

model.save(FINAL_MODEL_PATH)
print(f"\n  Final model → {FINAL_MODEL_PATH}")
print(f"  Best model  → {BEST_MODEL_PATH}")


# =============================================================================
# STEP 6: LOAD BEST MODEL
# =============================================================================

print("\n[STEP 6] Loading best checkpoint …")
best_model = tf.keras.models.load_model(
    BEST_MODEL_PATH,
    custom_objects={'EfficientNetPreprocess': EfficientNetPreprocess},
)
print("  ✅  Best model loaded successfully.")


# =============================================================================
# STEP 7: EVALUATION
# =============================================================================

print("\n[STEP 7] Generating Final Report Card …\n")

val_generator.reset()
y_pred_prob = best_model.predict(val_generator, verbose=1)
y_pred      = np.argmax(y_pred_prob, axis=1)
y_true      = val_generator.classes

print("\n--- Classification Report ---")
print(classification_report(y_true, y_pred, target_names=CLASS_NAMES, digits=4))

print("--- Per-class AUC (One-vs-Rest) ---")
print(f"  {'Class':<12} {'AUC':>8}  {'Note'}")
print(f"  {'─' * 42}")
for i, cls in enumerate(CLASS_NAMES):
    auc  = roc_auc_score((y_true == i).astype(int), y_pred_prob[:, i])
    note = "← most critical" if cls == 'malignant' else ""
    print(f"  {cls:<12} {auc:>8.4f}  {note}")

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES,
            linewidths=0.8, square=True)
plt.xlabel('AI Predicted Diagnosis', fontsize=12)
plt.ylabel('True Doctor Diagnosis',  fontsize=12)
plt.title('Confusion Matrix', fontsize=13, fontweight='bold')
plt.tight_layout(); plt.savefig('confusion_matrix.png', dpi=150); plt.show()

colors = ['#2196F3', '#F44336', '#4CAF50']
plt.figure(figsize=(8, 6))
for i, (cls, color) in enumerate(zip(CLASS_NAMES, colors)):
    bt = (y_true == i).astype(int)
    fpr, tpr, _ = roc_curve(bt, y_pred_prob[:, i])
    auc = roc_auc_score(bt, y_pred_prob[:, i])
    plt.plot(fpr, tpr, color=color, lw=2, label=f'{cls}  (AUC = {auc:.4f})')
plt.plot([0, 1], [0, 1], 'k--', lw=1, label='Random')
plt.xlabel('False Positive Rate', fontsize=12); plt.ylabel('True Positive Rate', fontsize=12)
plt.title('ROC Curves — One-vs-Rest', fontsize=13, fontweight='bold')
plt.legend(loc='lower right'); plt.grid(alpha=0.3)
plt.tight_layout(); plt.savefig('roc_curves.png', dpi=150); plt.show()

def _cat(key):
    return history_phase1.history.get(key, []) + history_phase2.history.get(key, [])

ep = list(range(1, len(_cat('loss')) + 1))
pb = len(history_phase1.history['loss'])
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Training History — Phase 1 + Phase 2', fontsize=14, fontweight='bold')
for ax, (tk, vk, label) in zip(axes.flat, [
    ('accuracy', 'val_accuracy', 'Accuracy'), ('loss', 'val_loss', 'Loss'),
    ('auc', 'val_auc', 'AUC'), ('recall', 'val_recall', 'Recall  ★ Priority'),
]):
    ax.plot(ep, _cat(tk), lw=2, label='Train')
    ax.plot(ep, _cat(vk), lw=2, linestyle='--', label='Validation')
    ax.axvline(pb, color='gray', linestyle=':', lw=1.5, label='Fine-tune ▶')
    ax.set_xlabel('Epoch'); ax.set_ylabel(label)
    ax.set_title(label, fontweight='bold'); ax.legend(fontsize=9); ax.grid(alpha=0.3)
plt.tight_layout(); plt.savefig('training_history.png', dpi=150); plt.show()


# =============================================================================
# STEP 8: GRAD-CAM  (v9 — definitive fix)
# =============================================================================
#
# ╔══════════════════════════════════════════════════════════════════╗
# ║  WHY ALL PREVIOUS APPROACHES FAILED                             ║
# ║                                                                  ║
# ║  v7: full_model.get_layer('top_conv')                           ║
# ║      → ValueError: top_conv not at top level                    ║
# ║                                                                  ║
# ║  v8: sub_model.get_layer('top_conv').output  used as output of  ║
# ║      a NEW Model(full_model.inputs, [conv.output, full.output]) ║
# ║      → KeyError: tensor ID not found during forward pass.       ║
# ║      After load_model(), the sub-model lives in its own graph   ║
# ║      scope. Its internal tensor IDs are not registered in       ║
# ║      full_model's graph. Keras cannot trace the path from       ║
# ║      full_model's input to a tensor inside the sub-model's      ║
# ║      separate graph → KeyError on tensor ID lookup.             ║
# ║                                                                  ║
# ║  v9: FORWARD HOOK PATTERN (no new Model construction)          ║
# ║  ─────────────────────────────────────────────────────          ║
# ║  Instead of building a new Model, we intercept the conv output  ║
# ║  DURING the existing forward pass of full_model:                ║
# ║                                                                  ║
# ║  1. Locate top_conv layer inside efficientnetb0 sub-model.      ║
# ║  2. Temporarily REPLACE its call method with a wrapper that     ║
# ║     saves the output to a Python list (the "hook").             ║
# ║  3. Run full_model(img) inside GradientTape — the wrapper fires ║
# ║     mid-pass and stores the conv tensor in the hook.            ║
# ║  4. Restore the original call method immediately after.         ║
# ║  5. Compute gradient of class_score w.r.t. the hooked tensor.  ║
# ║                                                                  ║
# ║  All tensors stay inside full_model's own forward pass —        ║
# ║  no graph boundaries are crossed, no new Model is built.        ║
# ║  GradientTape sees one connected graph → gradient flows ✅      ║
# ╚══════════════════════════════════════════════════════════════════╝

def make_gradcam_heatmap(
    img_array  : np.ndarray,
    full_model : tf.keras.Model,
) -> tuple:
    """
    Compute a Grad-CAM heatmap using a forward-hook pattern.

    Intercepts the last conv layer's output DURING full_model's own
    forward pass — no new Model is constructed, no graph boundaries
    are crossed. Works correctly after load_model() in TF 2.x.

    Parameters
    ----------
    img_array  : (1, H, W, 3) float32, pixels in [0, 255]
    full_model : Loaded Keras model (BreastCancer_EfficientNetB0_v9)

    Returns
    -------
    heatmap    : 2-D float32 array in [0, 1]
    pred_idx   : int   — predicted class index
    confidence : float — softmax probability of predicted class
    """
    # ── Step 1: Find the last conv layer inside the nested sub-model ──────────
    efficientnet_block = full_model.get_layer('efficientnetb0')
    target_layer       = None
    for layer in reversed(efficientnet_block.layers):
        if isinstance(layer, (layers.Conv2D, layers.DepthwiseConv2D)):
            target_layer = layer
            break
    if target_layer is None:
        raise ValueError("No Conv2D layer found inside efficientnetb0.")

    # ── Step 2: Register a forward hook ───────────────────────────────────────
    # We temporarily replace the layer's __call__ method with a wrapper
    # that captures the output tensor into a list during the forward pass.
    # Using a list (not a variable) so the tensor retains its gradient tape
    # connection — no copy is made, just a reference is stored.
    conv_output_hook = []
    original_call    = target_layer.__class__.call   # Save original method

    def hooked_call(self, inputs, **kwargs):
        output = original_call(self, inputs, **kwargs)
        if self is target_layer:          # Only hook our specific layer instance
            conv_output_hook.append(output)
        return output

    target_layer.__class__.call = hooked_call  # Monkey-patch the class call

    # ── Step 3: Single forward pass with GradientTape ─────────────────────────
    try:
        with tf.GradientTape() as tape:
            predictions = full_model(img_array, training=False)

            # The hook fired during the above call — retrieve the stored tensor
            if not conv_output_hook:
                raise RuntimeError("Forward hook did not capture conv output.")

            conv_outputs = conv_output_hook[0]  # Tensor from the actual forward pass
            tape.watch(conv_outputs)            # Tell tape to track this tensor

            pred_idx    = int(tf.argmax(predictions[0]))
            confidence  = float(predictions[0, pred_idx])
            class_score = predictions[:, pred_idx]

        # ── Step 4: Compute gradients ─────────────────────────────────────────
        grads = tape.gradient(class_score, conv_outputs)   # (1, h, w, C)

        if grads is None:
            raise RuntimeError("Gradient is None — tape lost track of conv_outputs.")

    finally:
        # ── Step 5: ALWAYS restore original call (even if an error occurs) ────
        target_layer.__class__.call = original_call
        conv_output_hook.clear()

    # ── Step 6: Build heatmap from gradients ──────────────────────────────────
    # Global average pooling over spatial dims → importance weight per channel
    channel_wts = tf.reduce_mean(grads, axis=(0, 1, 2))   # (C,)

    # Weighted sum of feature maps → 2-D spatial importance map
    conv_out = conv_outputs[0]                             # (h, w, C)
    heatmap  = tf.reduce_sum(channel_wts * conv_out, axis=-1).numpy()

    # ReLU (keep only positive activations) + normalise to [0, 1]
    heatmap = np.maximum(heatmap, 0)
    if heatmap.max() > 0:
        heatmap /= heatmap.max()

    return heatmap.astype(np.float32), pred_idx, confidence


def overlay_gradcam(
    original : np.ndarray,   # float32 [0, 1], (H, W, 3)
    heatmap  : np.ndarray,   # float32 [0, 1], (h, w)
    alpha    : float = 0.45,
) -> np.ndarray:
    """Resize heatmap to image size and blend with jet colormap."""
    h, w    = original.shape[:2]
    resized = cv2.resize(heatmap, (w, h))
    colored = cv2.applyColorMap((resized * 255).astype(np.uint8), cv2.COLORMAP_JET)
    colored = cv2.cvtColor(colored, cv2.COLOR_BGR2RGB).astype(np.float32) / 255.0
    blended = (1 - alpha) * original + alpha * colored
    return (blended.clip(0, 1) * 255).astype(np.uint8)


# ── Visualise 6 validation samples with Grad-CAM ─────────────────────────────
print("\n[STEP 8] Generating AI Explainability Heatmaps (Grad-CAM) …")

val_generator.reset()
sample_imgs, sample_labels = next(val_generator)
n_show = min(6, len(sample_imgs))

fig = plt.figure(figsize=(15, n_show * 3.8))
gs  = gridspec.GridSpec(n_show, 3, figure=fig, hspace=0.45, wspace=0.06)
fig.suptitle(
    "Grad-CAM — Which pixels did the AI look at?\n"
    "🔴 Red = most influential   🔵 Blue = ignored",
    fontsize=14, fontweight='bold', y=1.01,
)
col_headers = ["Original Scan", "Grad-CAM Heatmap", "Overlay + Verdict"]

for row in range(n_show):
    img_raw  = sample_imgs[row]
    img_disp = img_raw / 255.0
    img_in   = np.expand_dims(img_raw, axis=0)

    true_cls           = CLASS_NAMES[np.argmax(sample_labels[row])]
    heatmap, pred_idx, conf = make_gradcam_heatmap(img_in, best_model)
    pred_cls           = CLASS_NAMES[pred_idx]
    correct            = '✅' if pred_cls == true_cls else '❌'
    overlay            = overlay_gradcam(img_disp, heatmap)

    for col, (data, cmap) in enumerate([
        (img_disp, None), (heatmap, 'jet'), (overlay, None),
    ]):
        ax = fig.add_subplot(gs[row, col])
        ax.imshow(data, cmap=cmap)
        ax.axis('off')
        if row == 0:
            ax.set_title(col_headers[col], fontsize=11, fontweight='bold', pad=6)
        if col == 2:
            color = 'green' if pred_cls == true_cls else 'red'
            ax.set_xlabel(
                f"True: {true_cls.upper()}  |  AI: {pred_cls.upper()} {correct}  ({conf:.0%})",
                fontsize=9, color=color, labelpad=4,
            )

plt.savefig('gradcam_samples.png', dpi=150, bbox_inches='tight')
plt.show()
print("  ✅  Saved → gradcam_samples.png")


# =============================================================================
# BONUS: SINGLE-IMAGE DIAGNOSIS
# =============================================================================

def diagnose_image(
    image_path  : str,
    model       : tf.keras.Model = None,
    show_gradcam: bool           = True,
) -> dict:
    """
    Run AI diagnosis on any single ultrasound image.

    Parameters
    ----------
    image_path   : Path to .png / .jpg image
    model        : Trained model (defaults to best_model)
    show_gradcam : Show Grad-CAM overlay if True

    Returns
    -------
    dict — 'predicted_class', 'confidence', 'probabilities'

    Example
    -------
        result = diagnose_image("dataset/Dataset_BUSI_with_GT/malignant/malignant (1).png")
    """
    if model is None:
        model = best_model
    if not os.path.exists(image_path):
        raise FileNotFoundError(f"Not found: {image_path}")

    img  = tf.keras.utils.load_img(image_path, target_size=IMG_SIZE)
    arr  = tf.keras.utils.img_to_array(img)
    inp  = np.expand_dims(arr, axis=0)

    probs      = model.predict(inp, verbose=0)[0]
    pred_idx   = int(np.argmax(probs))
    pred_class = CLASS_NAMES[pred_idx]
    confidence = float(probs[pred_idx])

    result = {
        'predicted_class' : pred_class,
        'confidence'      : confidence,
        'probabilities'   : {c: float(p) for c, p in zip(CLASS_NAMES, probs)},
    }

    print(f"\n  {'─' * 44}")
    print(f"  🔬  AI Diagnosis Report")
    print(f"  {'─' * 44}")
    print(f"  File      : {os.path.basename(image_path)}")
    print(f"  Diagnosis : {pred_class.upper()}  ({confidence:.1%} confidence)\n")
    for cls, p in result['probabilities'].items():
        bar    = '█' * int(p * 28)
        marker = ' ← predicted' if cls == pred_class else ''
        print(f"  {cls:<12} {bar:<28} {p:.1%}{marker}")
    print(f"  {'─' * 44}")

    if show_gradcam:
        heatmap, _, _ = make_gradcam_heatmap(inp, model)
        disp = arr / 255.0
        ov   = overlay_gradcam(disp, heatmap)

        fig, axes = plt.subplots(1, 3, figsize=(13, 4))
        axes[0].imshow(disp);               axes[0].set_title('Original Scan');      axes[0].axis('off')
        axes[1].imshow(heatmap, cmap='jet'); axes[1].set_title('Grad-CAM Heatmap'); axes[1].axis('off')
        axes[2].imshow(ov)
        axes[2].set_title(f"AI: {pred_class.upper()}\n({confidence:.1%} confidence)")
        axes[2].axis('off')
        plt.suptitle("AI Diagnosis Explainability", fontsize=13, fontweight='bold')
        plt.tight_layout(); plt.show()

    return result


# Uncomment to test on any image:
# result = diagnose_image("dataset/Dataset_BUSI_with_GT/malignant/malignant (1).png")


# =============================================================================
# DONE
# =============================================================================

print("\n" + "=" * 60)
print("  🎉  Medical AI Lab Completed Successfully!")
print("=" * 60)
print(f"  Phase 1 best val_auc  :  {best_p1:.4f}")
print(f"  Phase 2 best val_auc  :  {best_p2:.4f}")
print()
print("  Saved files:")
for f in [BEST_MODEL_PATH, FINAL_MODEL_PATH,
          'confusion_matrix.png', 'roc_curves.png',
          'training_history.png', 'gradcam_samples.png']:
    exists = os.path.exists(f)
    kb     = os.path.getsize(f) / 1024 if exists else 0
    print(f"    {'✅' if exists else '❌'}  {f:<36} ({kb:>8.1f} KB)")
print()
print("  TensorBoard:")
print("    %load_ext tensorboard")
print("    %tensorboard --logdir logs")
print("=" * 60)
```
